```python

import asyncio

from dotenv import load_dotenv
import os

class PortfolioMonitor:
    def __init__(self):
        load_dotenv()  # Load environment variables from .env file
        self.host = os.getenv('IBKR_HOST')
        self.port = int(os.getenv('IBKR_PORT'))
        self.client_id = int(os.getenv('IBKR_CLIENT_ID'))
        self.client = None
        self.wrapper = None

    async def connect(self):
        self.wrapper = IBWrapper()
        self.client = IBClient(wrapper=self.wrapper)
        await self.client.connect_async(self.host, self.port, self.client_id)
        print("Connected to IBKR API")

    async def disconnect(self):
        await self.client.disconnect_async()
        print("Disconnected from IBKR API")

    async def fetch_portfolio(self):
        account_summary = await self.client.req_account_summary_async()
        print("Account Summary:")
        for summary in account_summary:
            print(summary)

        positions = await self.client.req_positions_async()
        print("Positions:")
        for position in positions:
            print(position)

    async def monitor_portfolio(self):
        self.wrapper.update_portfolio_event += self.on_portfolio_update

    async def on_portfolio_update(self, data):
        print("Portfolio Update:", data)

    async def start_monitoring(self):
        await self.connect()
        await self.fetch_portfolio()
        await self.monitor_portfolio()

    async def stop_monitoring(self):
        await self.disconnect()

# Example usage
async def main():
    monitor = PortfolioMonitor()
    await monitor.start_monitoring()
    # Run for a specified period to monitor portfolio updates
    await asyncio.sleep(60)
    await monitor.stop_monitoring()

if __name__ == "__main__":
    asyncio.run(main())
```
